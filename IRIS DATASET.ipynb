#### 9. Violin plot is used to visualize the distribution of daat and its probability distribution . this chart is a combination of a boxplot and a Density plot that is rotated and placed on each side , to show the distribution shape of the data. the thick black bar in the centre represents the interquartile range, the thin black line extendend from it represents the 95% confidence intervals , and the white dot is the median . Boxplots are limited in their display of the data , as their visual simplicity tends to hide significant details about how values in the data are distributed.

```python
vis6 = plt.gcf()
vis6.set_size_inches(10,7)
vis6 = sns. violinplot(data=iris , x='Species', y='SepalLengthCm')
```

## 18 . Stacked Histogram

```python
iris['Species'] = iris['Species'].astype('category')
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```

*** with stacked histogram we can see the distribution of Speal Length of Different Species together. this shows us the range of Speal Length for the three different Species of iris Flower.
